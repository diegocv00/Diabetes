**Distribución de diabéticos y no diabéticos según su nivel de IMC**

```python
def clasificar_imc(x):
    if x < 18.5:
        return "Bajo peso"
    elif 18.5 <= x < 24.9:
        return "Normal"
    elif 25 <= x < 29.9:
        return "Sobrepeso"
    else:
        return "Obesidad"
df['imc_category'] = df['bmi'].apply(clasificar_imc)
df_poseen_diabetes["imc_category"] = df_poseen_diabetes["bmi"].apply(clasificar_imc)
df_no_poseen_diabetes["imc_category"] = df_no_poseen_diabetes["bmi"].apply(clasificar_imc)
```

```python


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))

ax1.pie(df_no_poseen_diabetes['imc_category'].value_counts(), labels=df_no_poseen_diabetes['imc_category'].value_counts().index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
ax1.set_title("Distribución del índice de masa corporal (IMC) en pacientes sin diabetes")
counts_no_diabetes = df_no_poseen_diabetes['imc_category'].value_counts().to_dict()


ax2.pie(df_poseen_diabetes['imc_category'].value_counts(), labels=df_poseen_diabetes['imc_category'].value_counts().index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
ax2.set_title("Distribución del índice de masa corporal (IMC) en pacientes con diabetes")
counts_with_diabetes = df_poseen_diabetes['imc_category'].value_counts().to_dict()


plt.show()
```

Las personas con obesidad y diabetes es mayor en relación a las personas con obesidad y sin diabetes, esto indica que el IMC puede ser un factor influyente en el diagnóstico de diabetes

**Distribución de diabéticos y no diabéticos según su ritmo cardíaco**

```python
def clasificar_hr(x):
    if x < 60:
        return "Bradicardia"
    elif 60 <= x <= 100:
        return "Normal"
    else:
        return "Taquicardia"

df_poseen_diabetes["heart_rate_category"] = df_poseen_diabetes["heart_rate"].apply(clasificar_hr)
df_no_poseen_diabetes["heart_rate_category"] = df_no_poseen_diabetes["heart_rate"].apply(clasificar_hr)
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))


counts_no = df_no_poseen_diabetes['heart_rate_category'].value_counts()
bar1 = ax1.bar(counts_no.index, counts_no, color=sns.color_palette("pastel"))
ax1.set_title("Distribución de la frecuencia cardíaca en pacientes SIN diabetes")
ax1.set_xlabel("Frecuencia cardíaca")
ax1.set_ylabel("Cantidad de pacientes")
ax1.bar_label(bar1, padding=3)


counts_si = df_poseen_diabetes['heart_rate_category'].value_counts()
bar2 = ax2.bar(counts_si.index, counts_si, color=sns.color_palette("pastel"))
ax2.set_title("Distribución de la frecuencia cardíaca en pacientes CON diabetes")
ax2.set_xlabel("Frecuencia cardíaca")
ax2.set_ylabel("Cantidad de pacientes")
ax2.bar_label(bar2, padding=3)

plt.tight_layout() 
plt.show()
```

No se observa una gran relación entre una frecuencia cardíaca alta y el poseer o no diabetes. Además la mayoria de pacientes tiene una frecuencia cardiaca normal, entre 60-100 látidos

**Distribución de triglicéridos en personas con y sin diabetes**

```python

promedio_tg_no_diabetes = df_no_poseen_diabetes['triglycerides'].mean()
promedio_tg_con_diabetes = df_poseen_diabetes['triglycerides'].mean()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))

ax1.hist(df_no_poseen_diabetes['triglycerides'], bins=50, color='#8CFFDF')
ax1.set_xlabel("Triglicéridos (mg/dL)")
ax1.set_ylabel("Cantidad de pacientes")
ax1.set_title("Distribución de Triglicéridos en pacientes sin diabetes")
ax1.grid(axis='y', alpha=0.3)
ax1.axvline(promedio_tg_no_diabetes, color='black', linestyle='dashed', linewidth=2, label=f'Promedio: {promedio_tg_no_diabetes:.2f}')
ax1.text(promedio_tg_no_diabetes + 5, ax1.get_ylim()[1] * 0.9, f'Media: {promedio_tg_no_diabetes:.2f}', color='black', fontweight='bold')

ax2.hist(df_poseen_diabetes['triglycerides'], bins=50, color='#FFDD8C')
ax2.set_xlabel("Triglicéridos (mg/dL)")
ax2.set_ylabel("Cantidad de pacientes")
ax2.set_title("Distribución de Triglicéridos en pacientes con diabetes")
ax2.axvline(promedio_tg_con_diabetes, color='black', linestyle='dashed', linewidth=2)
ax2.text(promedio_tg_con_diabetes + 5, ax2.get_ylim()[1] * 0.9, f'Media: {promedio_tg_con_diabetes:.2f}', color='black', fontweight='bold')

print(f"Promedio de triglicéridos en pacientes sin diabetes: {promedio_tg_no_diabetes:.2f} mg/dL")
print(f"Promedio de triglicéridos en pacientes con diabetes: {promedio_tg_con_diabetes:.2f} mg/dL")

plt.show()

```

Los pacientes con diabetes presentan triglicéridos más elevados en promedio. Esto es un factor de riesgo cardiovascular importante asociado con la diabetes

**Distribución de glucosa postprandial en personas con y sin diabetes**

```python
def clasificar_glucosa_posprandial(x):
    if x < 140:
        return "Normal"
    elif 140 <= x <= 199:
        return "Prediabetes"
    else:
        return "Diabetes"

df_no_poseen_diabetes["glucose_postprandial_category"] = df_no_poseen_diabetes["glucose_postprandial"].apply(clasificar_glucosa_posprandial)
df_poseen_diabetes["glucose_postprandial_category"] = df_poseen_diabetes["glucose_postprandial"].apply(clasificar_glucosa_posprandial)
```

```python


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))


counts_no = df_no_poseen_diabetes['glucose_postprandial_category'].value_counts()
bar1 = ax1.bar(counts_no.index, counts_no, color=sns.color_palette("pastel"))
ax1.set_title("Distribución de la glucosa postprandial en pacientes sin diabetes")
ax1.set_xlabel("Glucosa postprandial (mg/dL)")
ax1.set_ylabel("Cantidad de pacientes")
ax1.bar_label(bar1, padding=3)


counts_si = df_poseen_diabetes['glucose_postprandial_category'].value_counts()
bar2 = ax2.bar(counts_si.index, counts_si, color=sns.color_palette("pastel"))
ax2.set_title("Distribución de la glucosa postprandial en pacientes con diabetes")
ax2.set_xlabel("Glucosa postprandial (mg/dL)")
ax2.set_ylabel("Cantidad de pacientes")
ax2.bar_label(bar2, padding=3)

plt.tight_layout() 
plt.show()

```

La glucosa postprandial (después de comer) es otro indicador importante. Los pacientes diabéticos muestran un control glucémico deficiente después de las comidas, con valores mucho más elevados que los no diabéticos

**Distribución del HbA1c (Hemoglobina Glicosilada) en personas con y sin diabetes**

```python
def clasificar_hba1c(x):
    if x < 5.7:
        return "Normal"
    elif 5.7 <= x <= 6.4:
        return "Prediabetes"
    else:
        return "Diabetes"

df_no_poseen_diabetes["hba1c_category"] = df_no_poseen_diabetes["hba1c"].apply(clasificar_hba1c)
df_poseen_diabetes["hba1c_category"] = df_poseen_diabetes["hba1c"].apply(clasificar_hba1c)
```

```python


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))

ax1.pie(df_no_poseen_diabetes['hba1c_category'].value_counts(), labels=df_no_poseen_diabetes['hba1c_category'].value_counts().index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
ax1.set_title("Distribución de nivel de HbA1c en pacientes sin diabetes")
counts_no_diabetes = df_no_poseen_diabetes['hba1c_category'].value_counts().to_dict()


ax2.pie(df_poseen_diabetes['hba1c_category'].value_counts(), labels=df_poseen_diabetes['hba1c_category'].value_counts().index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
ax2.set_title("Distribución de nivel de HbA1c en pacientes con diabetes")
counts_with_diabetes = df_poseen_diabetes['hba1c_category'].value_counts().to_dict()


plt.show()

```

El HbA1c es uno de los mejores indicadores del control glucémico a largo plazo. Es claramente visible que los pacientes con diabetes tienen valores de HbA1c significativamente más altos, reflejando un control glucémico deficiente. El valor de 6.5% es el umbral diagnóstico para diabetes

## **Resumen del análisis de variables metabólicas**

### Variables Analizadas:
1. **Triglicéridos**: Elevados en diabéticos → Factor de riesgo cardiovascular
2. **Glucosa en Ayunas**: Diferencia significativa entre grupos
3. **Glucosa Postprandial**: Control glucémico deficiente post-comidas en diabéticos
4. **Nivel de Insulina**: Muestra variabilidad interesante entre grupos
5. **HbA1c**: El indicador más discriminante - claramente separa ambos grupos

### Conclusiones Clave:
- HbA1c es el marcador más consistente para identificar diabetes
- Los niveles de glucosa en ayunas y postprandial son indicadores confiables
- Los triglicéridos elevados son un factor de comorbilidad importante en diabéticos
- La combinación de estos marcadores proporciona un diagnóstico más robusto
